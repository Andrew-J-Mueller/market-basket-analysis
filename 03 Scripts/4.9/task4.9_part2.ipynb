## 2.2 sampling data with seed(), line charts

* Set the seed for the random number generator.
* Create 32,404,859 random numbers (the length of your ords_prods_merge dataframe) ranging between 0 and 1.
* Code a logical test comparing each of those numbers to a value of 0.7 (either less than or greater than 0.7).
* Store the results of the test in a new list, dev, in the form of boolean values (True and False).

```python
#set seed for random number generator 
np.random.seed(4)
```

```python
#create boolean list 
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
#check, print dev
dev
```

```python
#store 70% of sample 
big = ords_prods_merge[dev]
```

```python
#store 30% of sample using tilde as False
small = ords_prods_merge[~dev]
```

```python
#check 
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
#set as dataframe
df_2 = small[['orders_day_of_week','prices']]
```

#line chart set as variable

```python
%%time
line_ordsdayweek_prices = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
#export chart
line_ordsdayweek_prices.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_ordsdayweek_prices.png'))
```
